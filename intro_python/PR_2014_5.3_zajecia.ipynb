```python
l =4
N =100
x = np.linspace(0,l,N)
y = np.linspace(0,l,N)
X,Y = np.meshgrid(x,y)
```

```python
f = np.sin(X**2+Y**2)
#f[ (X-2)**2+(Y-2)**2<1 ] = -1
plt.imshow(f)
```

```python
plt.plot(Y[:,12],f[:,12])
```
