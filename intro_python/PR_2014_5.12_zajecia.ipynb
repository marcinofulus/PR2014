```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

l = 4
N = 100
x = np.linspace(-l/2,l/2,N)
y = np.linspace(-l/2,l/2,N)
X,Y = np.meshgrid(x,y)
```

```python

dt =1e-4
Niter=1000
D=1
t = Niter*dt
P = 1/(4*np.pi*D*t)*np.exp(-(X**2+Y**2)/(4*D*t) )
```

```python
plt.imshow(P)
```

```python
plt.plot( X[50,:],P[50,:])
```

```python
u = np.zeros((N,N))
dx = l/ (N-1.)
u[N//2, N//2 ]= 1.0/dx**2
C = D*dt/ (dx**2)
print C
```

```python
for i in range(Niter):
    u[1:-1,1:-1] = u[1:-1,1:-1] +\
      C*(u[1:-1,:-2]+ u[1:-1,2:]+ u[:-2, 1:-1]+ u[2:,1:-1] -\
      4*u[1:-1,1:-1])
```

```python
plt.imshow(u)
```
